```python
#Form a data frame to analyze family size,ed, marital status, and wine purchases 
family_winepurchases_df = wine_sales_data_df[["Kidhome", "Teenhome", "marital_Together",
                                                      "marital_Widow", "MntWines", "marital_Divorced", "education_2n Cycle",
                                                      "education_Basic", "education_Graduation", "education_Master",
                                                      "education_PhD", "Income"]]
family_winepurchases_df.head(10)
```

```python
family_winepurchases_df.info()
```

```python
family_winepurchases_df.loc[:,"Kids"] = family_winepurchases_df.loc[:,"Kidhome"] + family_winepurchases_df.loc[:,"Teenhome"]
```

```python
family_winepurchases_df.drop(["Kidhome", "Teenhome"], axis=1, inplace=True)
```

```python
#Reorder columns
family_winepurchases_df = family_winepurchases_df[["marital_Together","marital_Widow",
                                                "marital_Divorced", "Kids",
                                                "education_2n Cycle","education_Basic", 
                                                "education_Graduation", "education_Master",
                                                "education_PhD", "Income", "MntWines"]]

family_winepurchases_df.head()
```

```python
#Sort by wine purchases
family_winepurchases_df.sort_values(by="MntWines", ascending=False)
```

```python
# Make a new DF that contains only parents
ParentsBuyWine_df = family_winepurchases_df[family_winepurchases_df.Kids > 0]
ParentsBuyWine_df.sort_values(by= "MntWines", ascending=False)
```

```python
# Rename the columns
ParentsBuyWine_df.rename(columns={"marital_Together": "Married", "marital_Widow": "Widowed",\
                                    "marital_Divorced": "Divorced", "education_2n Cycle": "2nd Cycle",\
                                    "education_Basic": "Basic", "education_Graduation": "UnderGrad",\
                                    "education_Master": "Master", "education_PhD": "PhD"}, inplace=True)
```

```python
#Combine 2nd cycle and master columns into a new column called College using .loc
ParentsBuyWine_df.loc[:,"Masters"] = ParentsBuyWine_df.loc[:,"2nd Cycle"] + ParentsBuyWine_df.loc[:,"Master"]
```

```python
#Remove 2nd cycle and Master from the data frame
ParentsBuyWine_df.drop(["2nd Cycle", "Master"], axis=1, inplace=True)
```

```python
# Create a customer id column in parntersbuywine_df as the first column in the data frame and assign id's to each row
ParentsBuyWine_df.insert(0, 'Customer ID', range(1, 1 + len(ParentsBuyWine_df)))
ParentsBuyWine_df.head()
```

```python
ParentsBuyWine_df.info()
```

```python
#Drop null values from data frame
ParentsBuyWine_df.dropna(inplace=True)
```

```python
#write a for loop to iterate through rows and remove 0 from
#columns masters, phd, basic, graduation, widowed, married, divorced
for index, row in ParentsBuyWine_df.iterrows():
    if row["Masters"] == 0:
        ParentsBuyWine_df.loc[index, "Masters"] = ""
    if row["PhD"] == 0:
        ParentsBuyWine_df.loc[index, "PhD"] = ""
    if row["Basic"] == 0:
        ParentsBuyWine_df.loc[index, "Basic"] = ""
    if row["UnderGrad"] == 0:
        ParentsBuyWine_df.loc[index, "UnderGrad"] = ""
    if row["Widowed"] == 0:
        ParentsBuyWine_df.loc[index, "Widowed"] = ""
    if row["Married"] ==0:
        ParentsBuyWine_df.loc[index, "Married"] = ""
    if row["Divorced"] == 0:
        ParentsBuyWine_df.loc[index, "Divorced"] =""
ParentsBuyWine_df.head()
```

```python
#Create a for loop for Basic, Graduation, Master, PhD replace 1 with the education level
#and then combine the columns into a new column called Education and drop the old columns
for index, row in ParentsBuyWine_df.iterrows():
    if row["Basic"] == 1:
        ParentsBuyWine_df.loc[index, "Basic"] = "Basic"
    if row["UnderGrad"] == 1:
        ParentsBuyWine_df.loc[index, "UnderGrad"] = "UnderGrad"
    if row["Masters"] == 1:
        ParentsBuyWine_df.loc[index, "Masters"] = "Masters"
    if row["PhD"] == 1:
        ParentsBuyWine_df.loc[index, "PhD"] = "PhD"
ParentsBuyWine_df["Education"] = ParentsBuyWine_df["Basic"] + ParentsBuyWine_df["UnderGrad"] + ParentsBuyWine_df["Masters"] + ParentsBuyWine_df["PhD"]
ParentsBuyWine_df.drop(["Basic", "UnderGrad", "Masters", "PhD"], axis=1, inplace=True)
ParentsBuyWine_df.head()
```

```python
#Create a for loop for Married, Divorced, Widowed replace 1 with the marital status and then combine the columns into a new column called Marital Status and drop the old columns
for index, row in ParentsBuyWine_df.iterrows():
    if row["Married"] == 1:
        ParentsBuyWine_df.loc[index, "Married"] = "Married"
    if row["Divorced"] == 1:
        ParentsBuyWine_df.loc[index, "Divorced"] = "Divorced"
    if row["Widowed"] == 1:
        ParentsBuyWine_df.loc[index, "Widowed"] = "Widowed"
ParentsBuyWine_df["Marital Status"] = ParentsBuyWine_df["Married"] + ParentsBuyWine_df["Divorced"] + ParentsBuyWine_df["Widowed"]
ParentsBuyWine_df.drop(["Married", "Divorced", "Widowed"], axis=1, inplace=True)
ParentsBuyWine_df.head()
```

```python
# Export DF
ParentsBuyWine_df.to_csv("ParentsBuyWine_df.csv", index=False)
```
